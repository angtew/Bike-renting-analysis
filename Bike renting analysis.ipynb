```python
#11.Create new categorical column called day_period. 0-6=night , 6-12=morning , 12-6=afternoon , 6-24=evening
```

```python

def categorize_hour(hour):
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

data['day_period'] = data['hour'].apply(categorize_hour)

print(data[['hour', 'day_period']].head())
```

```python
#5.Calculate total number of casual and registered bikes rental in years 2011,2012
```

```python
total_casual_2011 = data[data['year'] == 2011]['casual'].sum()
total_registered_2011 = data[data['year'] == 2011]['registered'].sum()

total_casual_2012 = data[data['year'] == 2012]['casual'].sum()
total_registered_2012 = data[data['year'] == 2012]['registered'].sum()

print(f"2011 - Total Casual: {total_casual_2011}, Total Registered: {total_registered_2011}")
print(f"2012 - Total Casual: {total_casual_2012}, Total Registered: {total_registered_2012}")
```

```python
#6.Calculate the mean of the hourly total rentals count by season.
```

```python
seasonal_mean = data.groupby('season')['count'].mean()

highest_mean_season = seasonal_mean.idxmax()
highest_mean_value = seasonal_mean.max()

print("Mean of hourly total rentals by season:")
print(seasonal_mean)
print(f"\nSeason with the highest mean: {highest_mean_season} (Mean: {highest_mean_value:.2f})")
```

```python
#7.Bikes rented by registered , Non-registred users - working days vs non-working days, 
```

```python
summary = data.groupby(['year', 'workingday'])[['registered', 'casual']].sum()

summary_2011 = summary.loc[2011]
summary_2012 = summary.loc[2012]

print("Summary for 2011:")
print(summary_2011)
print("\nSummary for 2012:")
print(summary_2012)

if summary_2011.loc[1, 'registered'] > summary_2011.loc[0, 'registered']:
    print("\nIn 2011, more bikes were rented by registered users on working days.")
else:
    print("\nIn 2011, more bikes were rented by registered users on non-working days.")

if summary_2012.loc[1, 'registered'] > summary_2012.loc[0, 'registered']:
    print("In 2012, more bikes were rented by registered users on working days.")
else:
    print("In 2012, more bikes were rented by registered users on non-working days.")


if summary_2011.loc[1, 'casual'] > summary_2011.loc[0, 'casual']:
    print("\nIn 2011, more bikes were rented by casual users on working days.")
else:
    print("\nIn 2011, more bikes were rented by casual users on non-working days.")

if summary_2012.loc[1, 'casual'] > summary_2012.loc[0, 'casual']:
    print("In 2012, more bikes were rented by casual users on working days.")
else:
    print("In 2012, more bikes were rented by casual users on non-working days.")
```

```python
#For "Registered users" , more bikes were rented on "working days".
#But for "Non-Registered users" , more bikes were rented on "non-working days".
#And they are the same for both years.
```

```python
#8.Months in 2011 & 2012 have the highest and lowest total number of bike rented
```

```python
data_2011 = data[data['year'] == 2011]

monthly_rentals = data_2011.groupby('month')['count'].sum()

highest_month = monthly_rentals.idxmax()
lowest_month = monthly_rentals.idxmin()

highest_rentals = monthly_rentals.max()
lowest_rentals = monthly_rentals.min()

print("Total rentals by month in 2011:")
print(monthly_rentals)
print(f"\nMonth with the highest rentals: {highest_month} ({highest_rentals} bikes)")
print(f"Month with the lowest rentals: {lowest_month} ({lowest_rentals} bikes)")
```

```python
data_2012 = data[data['year'] == 2012]
monthly_rentals = data_2012.groupby('month')['count'].sum()

highest_month = monthly_rentals.idxmax()
lowest_month = monthly_rentals.idxmin()

highest_rentals = monthly_rentals.max()
lowest_rentals = monthly_rentals.min()

print("Total rentals by month in 2012:")
print(monthly_rentals)
print(f"\nMonth with the highest rentals: {highest_month} ({highest_rentals} bikes)")
print(f"Month with the lowest rentals: {lowest_month} ({lowest_rentals} bikes)")
```

```python
#9 Mean of hourly total rentals count vs seasons
```

```python
season_month_means = data.groupby(['season', 'month'])['count'].mean().reset_index()

fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=True)

seasons = ['Spring', 'Summer', 'Fall', 'Winter']
colors = ['#4CAF50', '#FFC107', '#FF5722', '#2196F3'] 

for ax, season, color in zip(axes.flat, seasons, colors):
    season_data = season_month_means[season_month_means['season'] == season]
    
    ax.bar(season_data['month'], season_data['count'], color=color, alpha=0.8)
    
    ax.set_title(season, fontsize=14, fontweight='bold')
    ax.set_xlabel('Month')
    ax.set_ylabel('Mean Rentals Count')
    ax.set_xticks(range(1, 13))
    ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=45)

fig.tight_layout(pad=3.0)
fig.suptitle('Mean Hourly Total Rentals Count Across Months for Each Season', fontsize=16, fontweight='bold')

plt.show()
```

```python
#The rentals count is gradually raised from January and keep going up until June , it's continully high until September
#and then graudually drop from October on.
#We can assume that it is related to the weater as the weather is getting warmer during Spring and then
#It is the good weather from Summer untill the end of Fall , then goes down as the weather is getting colder in Winter.
```

```python
#11 Heatmap for correlation matrix of the dataset numerial variables
```

```python
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
numerical_data = data[numerical_columns]

correlation_matrix = numerical_data.corr()

plt.figure(figsize=(12, 10)) 
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)

plt.title('Correlation Matrix of Numerical Variables', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```

```python
#Strong Correlations with count (total bike rentals):
#registered users: There is a very strong positive correlation (close to 1.0) with count. This indicates that most bike rentals are from registered users.
#casual users: Also positively correlated, but less strongly than registered users.

#Weather and Temperature Variables:
#temperature: A positive correlation with count, meaning bike rentals tend to increase as the temperature rises.
#humidity: A weak or slightly negative correlation with count, suggesting that higher humidity might discourage rentals.
#windspeed: No strong correlation is visible with count, indicating windspeed does not significantly affect rentals.
#weather: A weak negative correlation, showing slightly fewer rentals under worse weather conditions.

#Working Day and Holiday:
#holiday: Negatively correlated with workingday (logical), but shows no significant correlation with count, indicating holidays don't strongly affect bike rentals.
#workingday: A weak positive correlation with count, suggesting more rentals occur on working days.

#Other Observations:
#temp and atemp (feels-like temperature): Strongly correlated with each other, as expected.
#humidity and temp: Weakly negative correlation, indicating hotter days tend to have lower humidity.
#windspeed and humidity: No strong relationship observed.

#Registered users are the dominant contributors to bike rentals.
#Bike rentals are positively influenced by temperature but are less affected by weather, humidity, and windspeed.
```
